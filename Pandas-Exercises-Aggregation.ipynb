### Step 4. Discover what is the mean age per occupation

```python
users.head(2)
users.groupby(users.occupation).age.mean().reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users['New Gender'] = np.where(users.gender == 'M',0,1)
users.head(2)
users.drop(columns=["New Gender"],axis=1,inplace=True)
users.head(2)
```

```python
users.head(2)
a = users[users.gender == 'M'].groupby(['occupation','gender']).user_id.count().reset_index()
a.sort_values(by="user_id",ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.head(2)
users.groupby(['occupation']).age.agg({min,max})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.head(2)
users.groupby(['occupation','gender']).age.mean().reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
a = 100 *users.groupby(["occupation","gender"]).user_id.count() / users.groupby(["occupation"]).user_id.count()
a.reset_index()
```
